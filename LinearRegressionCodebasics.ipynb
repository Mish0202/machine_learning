```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/sample_data/datapricearea.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price($US)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
reg.predict([[5300]])
```

```python
reg.coef_, reg.intercept_
```

```python
df_area = pd.read_csv("/content/sample_data/area.csv")
```

```python
df_area.head()
```

```python
prices = reg.predict(df_area)
```

```python
df_area['prices'] = prices
```

```python
df_area
```

```python
# @title area vs prices

from matplotlib import pyplot as plt
df_area.plot(kind='scatter', x='area', y='prices', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price($US)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
df_area.to_csv("predictedprices.csv")
```
